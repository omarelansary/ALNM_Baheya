Reading Data and Checking Columns of Missing Values

```python
data = pd.read_excel('cairo_final_usethis_6.xlsx')
```

```python
# data cleaning for grade column
grade_mapping = {'Low': 1, 'Intermediate': 2, 'High': 3}
data['Grade'] = data['Grade'].replace(grade_mapping)
data.Grade
```

```python
columns_to_remove = ['MRN', 'Birth.Date', 'First_BMI','Date.of.Diagnosis','Tumor size','Tumor size trim','Tumor_size_cm2','KI67%','Positive.LN','Total.LN.Dissected']
```

```python
categorical_data = data.drop(columns=columns_to_remove)
categorical_data
```

```python
# checking columns that have missing values or NULL or Unrecorded
missing_values = categorical_data.isnull().sum()
null_values = (categorical_data == "NULL").sum()
unrecorded_values = (categorical_data == "Unrecorded").sum()
missing_data=missing_values+unrecorded_values+null_values
missing_columns = missing_data[missing_data > 0]
print("Columns with missing values or 'NULL':")
print(missing_columns)
```

```python
categorical_data_missing=categorical_data[missing_columns.index]
categorical_data_missing_before_filling=categorical_data[missing_columns.index]
categorical_data_missing # Categorical Data Before Imputation
```

Mode Imputation For Categorical Data

```python
#Computing the mode for each column:
mode_values = categorical_data_missing.mode().iloc[0]
mode_values
```

```python
#Data Imputation using Mode for Missing Values:
for column in categorical_data_missing.columns:
      categorical_data_missing[column].fillna(mode_values[column], inplace=True)
```

```python
categorical_data_missing.head() #Categorical Data after imputation of missing cells
```

```python
# checking columns that have missing values or NULL or Unrecorded
missing_values2 = categorical_data_missing.isnull().sum()
null_values2 = (categorical_data_missing == "NULL").sum()
unrecorded_values2 = (categorical_data_missing == "Unrecorded").sum()
missing_data2=missing_values2+unrecorded_values2+null_values2
missing_columns2 = missing_data2[missing_data2 > 0]
print("Columns with missing values or 'NULL' or 'Unrecorded':")
print(missing_columns2)
```

```python
#Replacing unrecorded values in columns withs its mode:
categorical_data_missing['DM'].replace("Unrecorded", "No", inplace=True)
categorical_data_missing['HTN'].replace("Unrecorded", "No", inplace=True)
categorical_data_missing['VTE'].replace("Unrecorded", "No", inplace=True)
categorical_data_missing['CVD'].replace("Unrecorded", "No", inplace=True)
categorical_data_missing['F.Hx.of.BC'].replace("Unrecorded", "No", inplace=True)
categorical_data_missing['Menopausal.state'].replace("Unrecorded", "Post-M", inplace=True)
```

```python
categorical_data_missing #categorical data after filling missing values and unrecorded values
```

MICE  Imputer For Categorical Data

```python
categorical_data2 = data.drop(columns=columns_to_remove)
categorical_data_missing2=categorical_data2[missing_columns.index]
categorical_data_missing2
```

```python
categorical_data_encoded = pd.get_dummies(categorical_data_missing2)
categorical_data_encoded
```

```python
from sklearn.impute import IterativeImputer
```

```python
imputer = IterativeImputer(max_iter=10, random_state=0)
imputed_data = imputer.fit_transform(categorical_data_encoded)
```

```python
categorical_imputed = pd.DataFrame(imputed_data, columns=categorical_data_encoded.columns)
```
