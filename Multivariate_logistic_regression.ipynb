```python
import pandas as pd
import statsmodels.api as sm
```

```python
# Load your data into a pandas DataFrame (replace 'your_data.csv' with your actual file path)
data = pd.read_csv('cairo_bahya_numerical_cleaned.csv')
```

```python
# Convert relevant columns to numeric (if needed)
columns_to_convert = ['MRN','Unique MRN','First_BMI', 'Age.of.Diagnosis', 'KI67%', 'Grade', 'Positive.LN']
for col in columns_to_convert:
    data[col] = pd.to_numeric(data[col], errors='coerce')
```

```python
# Create a new DataFrame with the relevant columns
numeric_data = data[['First_BMI', 'Age.of.Diagnosis', 'KI67%', 'Grade', 'Total.LN.Dissected', 'Positive.LN']]
```

```python
# For example, let's fit a linear regression model:
X = numeric_data[['First_BMI', 'Age.of.Diagnosis', 'KI67%', 'Grade', 'Total.LN.Dissected']]
print(X.dtypes)
X = sm.add_constant(X)
y = numeric_data['Positive.LN']
```

```python
# Fit the linear regression model
model = sm.OLS(y, X).fit()
```

```python
# Get the summary of the model
print(model.summary())
```
