```python
import pandas as pd
import statsmodels.api as sm
```

```python
# Load your data into a pandas DataFrame (replace 'your_data.csv' with your actual file path)
data = pd.read_csv('cairo_bahya_numerical_cleaned.csv')
```

```python
# Specify the predictor variables (independent variables)
X = data[['First_BMI', 'Age.of.Diagnosis', 'KI67%', 'Grade', 'Total.LN.Dissected']]
```

```python
# Add a constant term for the intercept
X = sm.add_constant(X)
```

```python
# Specify the target variable (dependent variable)
y = data['Positive.LN']
```

```python
# Fit the linear regression model
model = sm.OLS(y, X).fit()
```

```python
# Get the summary of the model
print(model.summary())
```
